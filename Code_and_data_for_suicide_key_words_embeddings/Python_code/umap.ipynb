```python
import pandas as pd
import umap as umap
import matplotlib.pyplot as plt
```

```python
reducerm=umap.UMAP(random_state=0)
reducerf=umap.UMAP(random_state=0)
reducer=umap.UMAP(random_state=0, metric='manhattan', min_dist=0.00001)
```

```python
isuicd=pd.read_csv("~/JUDITA/mdpi_sustainability/suicide_coumtry_year_gender_bins.csv")
isuic=isuicd.set_index('country').drop(columns=['iso2c','iso3c'])
isuic.head()
```

```python
embeddingm=reducer.fit_transform(isuic)
plt.scatter(embeddingm[:,0], embeddingm[:,1])
plt.title("Both genders and distribution")
```

```python
df=pd.DataFrame(embeddingm, columns=['x','y'])
df['country']=isuicd['country']
df['iso2c']=isuicd['iso2c']
df['iso3c']=isuicd['iso3c']
print(df.head() )
df.to_csv("~/JUDITA/mdpi_sustainability/umap.csv")
```

```python
isuicm=pd.read_csv("salys_vyrai.csv")
isuicm=isuicm.set_index('Country')
isuicm.head()
```

```python
isuicf=pd.read_csv("salys_moterys.csv")
isuicf=isuicf.set_index('Country')
isuicf.head()
```

```python
embeddingm=reducerm.fit_transform(isuicm)
plt.scatter(embeddingm[:,0], embeddingm[:,1])
plt.title("Males")
```

```python
embeddingf=reducerf.fit_transform(isuicf)
plt.scatter(embeddingf[:,0], embeddingf[:,1])
plt.title("Females")
```

```python
umapm=pd.DataFrame(data=embeddingm, index=isuicm.index, columns=['males_x','males_y'])
umapm.head()
```

```python
umapf=pd.DataFrame(data=embeddingf, index=isuicf.index, columns=['females_x','females_y'])
umapf.head()
```

```python
umap_coord=pd.DataFrame.merge(umapm, umapf, left_index=True, right_index=True)
umap_coord.head()
```

```python
umap_coord.to_csv("salys_umap_coord.csv")
```
