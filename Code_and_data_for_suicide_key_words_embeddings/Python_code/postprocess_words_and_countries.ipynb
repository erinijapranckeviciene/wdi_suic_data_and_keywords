```python
columns=df.columns
print(columns)
```

```python
import numpy as np

elements=[]

for i in df.index:
    #print(i)
    v=df.iloc[i]
    for j, country in enumerate(columns[:-3]):
        element={}
        element['keyword']=v[-1]
        element['country']=columns[j]
        element['kwd_in_country']=v[j]
        element['kwd_total']=v[-2]
        element['countries_with_kwd']=v[-3]
        elements.append(element) 
        
words_and_countries=pd.DataFrame(elements)
```

```python
print(words_and_countries.shape)
print(words_and_countries.head)
```

```python
words_and_countries.to_csv("words_and_countries_list.csv", index=False)
```

## Find keywords only specific to country, where countries with kwd ==1

```python
country_specific_kwds=words_and_countries[ (words_and_countries['countries_with_kwd'] == 1) & (words_and_countries['kwd_in_country'] >0 ) ]  
print(country_specific_kwds.shape)
print( country_specific_kwds.head)
```

```python
N=10
country_specific_kwds_top10=country_specific_kwds.groupby('country').apply(lambda x : x.sort_values(by='kwd_in_country', ascending=False).head(N).reset_index(drop=True))
country_specific_kwds_top10_2=country_specific_kwds.groupby('country').apply(lambda x : x.sort_values(by='kwd_in_country', ascending=False).head(N))

print(country_specific_kwds_top10.head)
country_specific_kwds_top10.to_csv("country_specific_kwds_top10.csv" , index=False)
```

```python
specific_countries, counts=np.unique(country_specific_kwds['country'] , return_counts=True)
sc=pd.DataFrame({'specific_countries':specific_countries, 'counts':counts}).sort_values(by='counts', ascending=False)
print(sc.head(n=60))
# compute keywords - take first 10 most frequent keywords DONE
sc.to_csv("country_specific_kwd_counts_per_country.csv", index=False)
```
